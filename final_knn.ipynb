```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("archive/breast-cancer.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
# getting null values
df.isna().sum()
```

```python
null_columns =["fractal_dimension_mean","compactness_se","perimeter_mean"]

for col in null_columns:
    df[col] =  df[col].fillna(df[col].mean())

df.isna().sum()
```

```python
df.info()
```

```python
df.head(20)
```

```python
df.drop('id',axis=1,inplace=True)
df.head()
```

```python
x= df.drop('diagnosis',axis=1)
x.shape
```

```python
y=df['diagnosis']
y.shape
```

```python
y=y.map({'B':0,'M':1})
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.2, random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_classifier = KNeighborsClassifier(n_neighbors=5)
knn_classifier.fit(x_train,y_train)
```

```python
y_pred = knn_classifier.predict(x_test)
```

```python
y_pred
```

```python
accuracy = (y_test == y_pred).sum() / len(y_test)
print(accuracy)
```

```python
from sklearn.metrics import accuracy_score

print("accuracy for k=5 is", accuracy_score(y_test,y_pred))
```

```python
# finding best k

accuracies =[]
for k in range(1,21):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(x_train,y_train)
    #acc = model.score(x_test, y_test)  # 
    acc = accuracy_score(y_test, model.predict(x_test))
    accuracies.append(acc)

```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(1, 21), accuracies, marker='o', color='blue', label='Test Accuracy')
plt.xlabel('k - Number of Neighbors')
plt.ylabel('Accuracy')
plt.title('Accuracy vs. k in KNN')
plt.xticks(range(1, 21))
plt.grid(True)
plt.legend()
plt.show()
```

```python
best_k = accuracies.index(max(accuracies)) + 1
best_k
```

```python
from sklearn.model_selection import train_test_split, cross_val_score

cv_scores = []
k_range = range(1, 21)

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, x_train, y_train, cv=5, scoring='accuracy')
    cv_scores.append(scores.mean())

# Plotting Accuracy vs k
plt.figure(figsize=(10, 5))
plt.plot(k_range, cv_scores, marker='o', color='green', label='CV Accuracy')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Cross-Validation Accuracy')
plt.title('Finding the Best k (Using 5-Fold CV)')
plt.xticks(k_range)
plt.grid(True)
plt.legend()
plt.show()

# Best k
best_k = k_range[np.argmax(cv_scores)]
print(f"\nBest k found using cross-validation: {best_k} with accuracy = {max(cv_scores):.4f}")

```

```python
final_model = KNeighborsClassifier(n_neighbors=best_k)
final_model.fit(x_train, y_train)

# --------------------------------------------
# Step 9: Evaluate on Unseen Test Set
# --------------------------------------------
y_pred = final_model.predict(x_test)

print("\nFinal Evaluation on Test Set")
print("Accuracy:", accuracy_score(y_test, y_pred))
```
